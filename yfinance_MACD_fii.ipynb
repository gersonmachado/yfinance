```python
import pandas as pd
import yfinance as yf
import numpy as np
ignore_warnings = True
```

```python
#Escolher ação

Ticker = yf.Ticker('VINO11.SA')

#Escolher intervalo de dados
df = Ticker.history(period='3mo')
df = df['Close']

```

```python
df_ticker_daily = Ticker.history(period='3mo', interval='5m')
# períodos ['1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max']
# intervalos [“1m”, “2m”, “5m”, “15m”, “30m”, “60m”, “90m”, “1h”, “1d”, “5d”, “1wk”, “1mo”, “3mo”]
df_ticker_daily = df_ticker_daily['Close']
```

```python
#Transformar em df
df = pd.DataFrame(df)
df_ticker_daily = pd.DataFrame(df_ticker_daily)
```

```python
df_ticker_daily.reset_index(inplace=True)
df.reset_index(inplace=True)
```

```python
df
```

```python
# Novo df com dados históricos
df_ticker = df.copy()
```

```python
# Remove a última data do df Daily
df_remove_ticket = df_ticker.loc[(df_ticker['Date'] == pd.to_datetime('today').normalize())]
df_ticker = df_ticker.drop(df_remove_ticket.index)
df_ticker
```

```python
# Pegar o último preço negociado
df_ticker_daily_last_price = df_ticker_daily.tail(1)
```

```python
df_ticker_daily_last_price
```

```python
#Ajustar o formato da data e nome coluna
df_ticker_daily_last_price.rename(columns={'Datetime': 'Date'}, inplace=True)
df_ticker_daily_last_price
```

```python
df_ticker_daily_last_price_formated = pd.to_datetime(df_ticker_daily_last_price['Date'], format='%Y-%m-%d')
```

```python
#Append data atual
# df_ticker = df_ticker.append(df_ticker_daily_last_price) erro, usar concat
df_ticker = pd.concat([df_ticker, df_ticker_daily_last_price])
df_ticker
```

```python
#Ajustar data
df_ticker['Date'] = pd.to_datetime(df_ticker['Date'], utc=True).dt.date
df_ticker
```

```python
# Calculo do MACD -- Estudar sobre
# Função para calcular a média móvel exponencial (EMA)
rapidaMME = df_ticker.Close.ewm(span=12).mean()
lentaMME = df_ticker.Close.ewm(span=26).mean()
MACD = rapidaMME - lentaMME
sinal = MACD.ewm(span=9).mean()

df_ticker['MACD'] = MACD
df_ticker['Sinal'] = sinal
df_ticker
```

```python
#ajuste de index e retira data pregão

df_ticker = df_ticker.set_index(pd.DatetimeIndex(df_ticker['Date'].values))
df_ticker = df_ticker.drop(['Date'], axis=1)
df_ticker
```

```python
# Criar código para verificar compra ou venda
df_ticker['flag'] = ''
df_ticker['preco_compra']=np.nan
df_ticker['preco_venda']=np.nan


for i in range (1, len(df_ticker.Sinal)):
  if df_ticker['MACD'][i] > df_ticker['Sinal'][i]:
    if df_ticker['flag'][i-1] == 'V':
      df_ticker['flag'][i] = 'V'
    else:
      df_ticker['flag'][i] = 'V'
      df_ticker['preco_venda'][i] = df_ticker['Close'][i]

  elif df_ticker['MACD'][i] < df_ticker['Sinal'][i]:
    if df_ticker['flag'][i-1] == 'C':
      df_ticker['flag'][i] = 'C'
    else:
      df_ticker['flag'][i] = 'C'
      df_ticker['preco_compra'][i] = df_ticker['Close'][i]

df_ticker
```

```python
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots



# Para buy and hold abaixando PM

df_plot = df_ticker.copy()

fig = go.Figure()

fig.add_trace(go.Scatter(x=df_plot.index,
                         y=df_plot['Close'],
                         name='Preço fechamento',
                         line_color='#FECB52'
                         ))
fig.add_trace(go.Scatter(x=df_plot.index,
                         y=df_plot['preco_venda'],
                         name='Venda',
                         mode= 'markers',
                         marker=dict(color='#EF553B', size=10,)
                         ))
fig.add_trace(go.Scatter(x=df_plot.index,
                         y=df_plot['preco_compra'],
                         name='Compra',
                         mode= 'markers',
                         marker=dict(color='#00CC96', size=10,)
                         ))

fig.show()

```
